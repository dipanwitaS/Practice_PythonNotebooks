```python
# Replacing all Nan values by 0
df['Temperature (C)'] = df['Temperature (C)'].fillna(0).astype(np.int64, errors='ignore')
df['Apparent Temperature (C)'] = df['Apparent Temperature (C)'].fillna(0).astype(np.int64, errors='ignore')
df['Humidity'] = df['Humidity'].fillna(0).astype(np.int64, errors='ignore')
df['Wind Speed (km/h)'] = df['Wind Speed (km/h)'].fillna(0).astype(np.int64, errors='ignore')
df['Wind Bearing (degrees)'] = df['Wind Bearing (degrees)'].fillna(0).astype(np.int64, errors='ignore')
df['Visibility (km)'] = df['Visibility (km)'].fillna(0).astype(np.int64, errors='ignore')
df['Pressure (millibars)'] = df['Pressure (millibars)'].fillna(0).astype(np.int64, errors='ignore')

```

```python
# Replacing all 0 values by mean
df['Temperature (C)'] = df['Temperature (C)'].fillna((df['Temperature (C)'].mean()))
df['Apparent Temperature (C)'] = df['Apparent Temperature (C)'].fillna((df['Apparent Temperature (C)'].mean()))
df['Humidity'] = df['Humidity'].fillna((df['Humidity'].mean()))
df['Wind Speed (km/h)'] = df['Wind Speed (km/h)'].fillna((df['Wind Speed (km/h)'].mean()))
df['Wind Bearing (degrees)'] = df['Wind Bearing (degrees)'].fillna((df['Wind Bearing (degrees)'].mean()))
df['Visibility (km)'] = df['Visibility (km)'].fillna((df['Visibility (km)'].mean()))
df['Pressure (millibars)'] = df['Pressure (millibars)'].fillna((df['Pressure (millibars)'].mean()))
```

# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df4,y, test_size = 0.3)
```

## Training the Model


```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
rfc.fit(x_train,y_train.values.ravel())
```

```python
rfc.get_params()
```

```python
y_pred = rfc.predict(x_test)
```

```python
## Accuracy check
```

```python
from sklearn import metrics
```

```python
print('Accuracy of model is ',metrics.accuracy_score(y_test,y_pred))
```

```python
feature_imp = pd.Series(rfc.feature_importances_, index= df4.columns)
```

```python
feature_imp
```

```python
sns.barplot(x=feature_imp,y=feature_imp.index)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
# fit the model 
knn.fit(x_train,y_train.values.ravel())
```

```python
pred = knn.predict(x_test)
```

```python

from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
